```python
import numpy as np
import random 
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(5,1,1000)
y = x*3.4 + 2.65 + np.random.randint(-8,8,1000)
plt.scatter(x,y)
```

```python
x_train = x[0:800]
x_test = x[800:]
y_train = y[0:800]
y_test = y[800:]
```

```python
def linear_regression(x,y):
    x_mean = x.mean()
    y_mean = y.mean()
    
    intercept = (np.multiply(x,y) - np.multiply(y_mean,x)).sum()/(x**2 - np.multiply(x_mean,x)).sum()
    slope = y_mean - intercept*x_mean
    
    return lambda x: slope*x + intercept
```

```python
func = linear_regression(x_train,y_train)
```

```python
func(x_train[0])
```

```python
y_pred = func(x_test)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,func(x_train),color = "r")
plt.scatter(x_test,y_pred,color = "purple")
```
